```python
import yaml

# Define the base path
base_path = '/kaggle/working/License-Plate-Detection-1/'

# Path to the data.yaml file
file_path = '/kaggle/working/License-Plate-Detection-1/data.yaml'

# Load the existing YAML file
with open(file_path, 'r') as file:
    data = yaml.safe_load(file)

# Update the paths
data['train'] = f'{base_path}/{data["train"]}'
data['val'] = f'{base_path}/{data["val"]}'
data['test'] = f'{base_path}/{data["test"]}'

# Save the updated YAML file
with open(file_path, 'w') as file:
    yaml.safe_dump(data, file)

print(f"Modified YAML file saved to: {file_path}")
```
